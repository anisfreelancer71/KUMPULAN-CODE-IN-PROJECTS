```python
import pandas as pd

# Baca file CSV menjadi DataFrame
data1 = pd.read_csv(file_path1)
data2 = pd.read_csv(file_path2)
data3 = pd.read_csv(file_path3)
data4 = pd.read_csv(file_path4)

# Gabungkan DataFrame menggunakan pd.concat
data_gab = pd.concat([data1, data2, data3, data4], ignore_index=True)

# Simpan DataFrame gabungan ke file CSV
data_gab.to_csv('combined.csv', index=False)

# Cetak DataFrame gabungan
print(data_gab)
```

```python
# Menghapus kolom-kolom "Unnamed: 8" hingga "Unnamed: 1000" dari DataFrame
columns_to_drop = [f'Unnamed: {i}' for i in range(8, 1001)]
data_gab.drop(columns=columns_to_drop, inplace=True)
```

```python
data_gab
```

```python
# Mengisi nilai NaN dengan nilai mean dari masing-masing kolom
data_gab.fillna(data_gab.mean(), inplace=True)

# Simpan DataFrame gabungan dengan nilai NaN yang telah diisi ke file CSV
data_gab.to_csv('combined_with_filled_nan.csv', index=False)

# Cetak DataFrame gabungan
print(data_gab)
```

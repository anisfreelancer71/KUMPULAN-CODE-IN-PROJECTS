```python
# Impor pustaka yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari file CSV
df = pd.read_csv("CreditCardClients.csv")

# Contoh plot dengan Matplotlib
# Misalnya, histogram untuk kolom "LIMIT_BAL"
plt.figure(figsize=(8, 6))
plt.hist(df['LIMIT_BAL'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribusi Limit Balance')
plt.xlabel('Limit Balance')
plt.ylabel('Frekuensi')
plt.show()

# Contoh plot dengan Seaborn
# Misalnya, boxplot untuk melihat distribusi "AGE" berdasarkan "EDUCATION"
plt.figure(figsize=(10, 6))
sns.boxplot(x='EDUCATION', y='AGE', data=df, palette='pastel')
plt.title('Distribusi Usia Berdasarkan Pendidikan')
plt.xlabel('Pendidikan')
plt.ylabel('Usia')
plt.show()
```

## Scatter Plot

```python
# Scatter plot antara "LIMIT_BAL" dan "AGE"
plt.figure(figsize=(8, 6))
plt.scatter(df['AGE'], df['LIMIT_BAL'], alpha=0.5, color='green')
plt.title('Scatter Plot antara Usia dan Limit Balance')
plt.xlabel('Usia')
plt.ylabel('Limit Balance')
plt.show()
```

##Count Plot dengan Seaborn

```python
# Count plot untuk kolom "EDUCATION"
plt.figure(figsize=(8, 6))
sns.countplot(x='EDUCATION', data=df, palette='Set1')
plt.title('Count Plot Pendidikan')
plt.xlabel('Pendidikan')
plt.ylabel('Frekuensi')
plt.show()
```

##Heatmap Korelasi dengan Seaborn

```python
# Heatmap untuk melihat korelasi antar kolom
correlation_matrix = df.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap korelasi antar kolom'.title())  # Mengubah judul menjadi huruf kecil
plt.show()

```

#Pair Plot dengan Seaborn

```python
# Pair plot untuk melihat hubungan antara beberapa kolom numerik
sns.pairplot(df[['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1', 'default.payment.next.month']], hue='default.payment.next.month', palette='husl')
plt.title('Pair Plot')
plt.show()
```

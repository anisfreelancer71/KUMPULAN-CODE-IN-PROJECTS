```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Membaca data dari file CSV
dataku = pd.read_csv('/content/drive/My Drive/filecsv/shopping_data.csv')
dataku_df = pd.DataFrame(dataku)

# Mengubah kolom Genre menjadi biner
dataku_df['Genre'] = dataku_df['Genre'].replace({'Female': 0, 'Male': 1})

# Membuat histogram dengan warna berbeda untuk nilai biner (0 dan 1)
plt.hist([dataku_df[dataku_df['Genre'] == 0]['Genre'], dataku_df[dataku_df['Genre'] == 1]['Genre']],
         bins=2, color=['red', 'blue'], label=['Female', 'Male'])

plt.title('Distribusi Genre per Gender')
plt.xlabel('Genre')
plt.ylabel('Frekuensi')
plt.grid(True)

plt.legend()
plt.show()

```

```python
# Menghitung jumlah Female dan Male dalam kolom 'Genre'
genre_counts = dataku_df['Genre'].value_counts()

# Mencetak hasil
print("Female =", genre_counts[0])
print("Male =", genre_counts[1])
```

```python
# Mengelompokkan berdasarkan nilai 'Genre' dan menghitung jumlahnya
genre_counts = dataku_df.groupby('Genre').count()

# Mencetak hasil
print("Female =", genre_counts.loc[0, 'Age'])
print("Male =", genre_counts.loc[1, 'Age'])
```

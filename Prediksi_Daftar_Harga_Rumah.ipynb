```python
# Import library yang dibutuhkan
import pandas as pd
from sklearn.linear_model import LinearRegression

# Membaca data dari file CSV
file_path = '/content/drive/My Drive/filecsv/data_harga_rumah.csv'
data = pd.read_csv(file_path)

# Menghapus kolom yang tidak diperlukan
data.drop(['NAMA RUMAH'], axis=1, inplace=True)
kolom_hapus = [f'Unnamed: {i}' for i in range(8, 1001)]
data.drop(columns=kolom_hapus, inplace=True)

# Memilih fitur dan target
X = data[['LB', 'LT', 'KT', 'KM', 'GRS']]  # Fitur
y = data['HARGA']  # Target

# Membagi data menjadi data pelatihan dan data pengujian
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# Melatih model regresi linear
model = LinearRegression()
model.fit(X_train, y_train)

# Input pengguna
print("Masukkan informasi rumah:")
LB = float(input("Luas Bangunan (LB): "))
LT = float(input("Luas Tanah (LT): "))
KT = int(input("Jumlah Kamar Tidur (KT): "))
KM = int(input("Jumlah Kamar Mandi (KM): "))
GRS = int(input("Jumlah Garasi (GRS): "))

# Membuat prediksi berdasarkan input pengguna
predicted_price = model.predict([[LB, LT, KT, KM, GRS]])

# Menampilkan hasil prediksi dalam format Rupiah
print(f"Prediksi Harga Rumah: RP {predicted_price[0]:,.2f}")
from ipywidgets import interact

@interact(LB=(0, 500), LT=(0, 1000), KT=(1, 10), KM=(1, 10), GRS=(0, 5))
def predict_price(LB=[], LT=[], KT=[], KM=2, GRS=[]):
    # Kode prediksi harga berdasarkan input pengguna
    predicted_price = model.predict([[LB, LT, KT, KM, GRS]])
    print(f"Prediksi Harga Rumah: RP {predicted_price[0]:,.2f}")
```

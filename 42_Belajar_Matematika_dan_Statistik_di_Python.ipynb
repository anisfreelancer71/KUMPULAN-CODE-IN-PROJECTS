```python
# Mengiport library
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
sns.set()
```

```python
# Menentukan rataan 2 kelompok
# group 1 = tanpa pupuk
# group 2 = dengan pupuk
rataan1 = 50
rataan2 = 65

# Menentukan standard deviasi(variasi) -- sama untuk dua kelompok
sd = 5

# Ukuran sample dua kelompok
sample1 = 40
sample2 = 50
```

```python
# Generate bilangan random
test = np.random.randn(sample1)
test
```

```python
plt.plot(test);
```

```python
sns.histplot(test);
```

```python
# Generate data
data1 = np.random.randn(sample1)*sd + rataan1
data2 = np.random.randn(sample2)*sd + rataan2

# Menggabungkan sample size (jumlah data)
kelompok_data = [sample1, sample2]

batas = [np.min(np.hstack((data1, data2))), np.max(np.hstack((data1, data2)))]
```

```python
batas
```

```python
# Plot distribusi kedua kelompok (statistik deskriptif)
plt.figure(figsize=(7,5))

a = sns.histplot(data1, label='data1')
a = sns.histplot(data2, label='data2')

a.set(xlabel='Data', ylabel='Density')
plt.legend()
plt.show()
```

```python
# Melakukan uji statistik (statistik inferential)
fig, ax = plt.subplots(1,2, figsize=(7,4), dpi=200)

ax[0].violinplot(data1)
ax[0].plot(1+np.random.randn(sample1)/5, data1, color='green')
ax[0].set_ylim(batas)

ax[1].violinplot(data2)
ax[1].plot(1+np.random.randn(sample2)/5, data2, color='blue')
ax[1].set_ylim(batas)
# Melakukan Uji test
t,p = stats.ttest_ind(data1, data2)

# Mencetak hasil t-test di judul
sigtxt = ('', ' TIDAK')
plt.title('Dua kelompok{} berbeda secara signifikan! t({})={}, p ={}'.format(sigtxt[int(p>.05)],
                                                                             sum(kelompok_data)-2,
                                                                             np.round(t,2),
                                                                             np.round(p,6)))
plt.show()
```
